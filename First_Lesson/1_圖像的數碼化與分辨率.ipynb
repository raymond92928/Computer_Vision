```python
import cv2
import sys
import plotly.express as px
```

```python
path = "/content/Computer_Vision/First_Lesson/images/yh_black.jpg" # 設定影像路徑
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE) # 以灰階圖模式讀取影像，黑白影像是特殊的灰階圖
# 顯示影像
'''
 img參數為影像對象
 color_continuous_scale 為顯示影像的色階，gray指與黑白漸進的灰階顯示
 zmin設定影像色階的最小值
 zmax設定影像色階的最大值
'''
fig = px.imshow(img,
        color_continuous_scale='gray',
        zmin = 0,
        zmax = 255)
fig.show()
```

運行上方程序，在視覺化視窗中移動滑鼠指針，可以查看影像中任意位置的像素值，其中：
* x代表選取像素點橫向座標，即第x列
* y代表選取像素點縱向座標，即第y行
* color代表選取像素點(x, y)的像素值，在黑白圖中，像素值只有0與255兩個取值

提示：
CV2讀取影像時，顏色通道為GBR，為了與原始圖片的RGB通道同步，需要轉換顏色通道，`img = img[:,:,::-1]` 便能達到此效果。

```python
path = "/content/Computer_Vision/First_Lesson/images/yh_color.jpg" # 設定圖片路徑
img = cv2.imread(path)[:, :, ::-1] # 讀取影像，並進行通道翻轉

# 顯示影像
'''
 img參數為影像對象
 zmin設定顯示影像時各通道色階的最小值
 zmax設定顯示影像時各通道色階的最大值
'''
fig = px.imshow(img,
        zmin = 0,
        zmax = 255)
fig.show()
```

運行上方程序，在視覺化視窗中移動滑鼠指針，可以查看影像中任意位置的像素值，其中：
* x代表選取像素點橫向座標，即第x列
* y代表選取像素點縱向座標，即第y行
* color代表選取像素點(x, y)處的像素值，分別為（R通道，G通道，B通道）

**【教師示範】檢視並修改影像解析度**

1. 載入任意圖片
2. 查看影像的解析度尺寸（shape）與寬高

備註：影像物件的shape中第一個元素是影像的高度（像素行數），第二個尺寸是影像的寬度（像素列數）

```python
img = cv2.imread("/content/Computer_Vision/First_Lesson/images/yh_color.jpg") # 讀入映像
img_ = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # 另一種轉換影像通道的方式
height, width= img_.shape[:2] # 讀取影像的尺寸，並將前兩個回傳值分別存入width,height

print("查看圖片的尺寸：", img_.shape[:2])
print("影像的寬度為（即像素格子的列數）：", width)
print("影像的高度為（即像素格子的行數）：", height)
```

3. 修改影像的解析度尺寸

```python
# 修改圖片尺寸
new_img = cv2.resize(img_, (width // 20, height // 20)) # 以第二個參數指定的數值修改圖片尺寸

print("查看新圖片的尺寸：", new_img.shape[:2])
```

4. 顯示修改前與修改後的影像

註：後續選擇較常用的matplotlib進行影像視覺化

```python
import matplotlib.pyplot as plt

ax1 = plt.subplot(1,2,1)

plt.imshow(img_)
ax2 = plt.subplot(1,2,2)

plt.imshow(new_img)
plt.show() # 同時顯示原始影像與縮小後的影像
```
