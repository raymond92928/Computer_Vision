### **[教師示範] RGB影像的冷色調效果**

1. 載入必須的第三方函式庫

```python
import cv2
import matplotlib.pyplot as plt
```

2. 讀取彩色影像 `yh_color.jpg`.

```python
img = cv2.imread("/content/Computer_Vision/First_Lesson/images/yh_color.jpg")[:, :, ::-1]
```

3. 將影像分割為 3 個色彩通道。

`cv2.split()` 函數傳回 3 個數列，每個數列分別代表紅色、綠色和藍色通道。

```python
r, g, b = cv2.split(img)
```

4. 以兩種不同的方式應用冷色調效果：

 a. 增加藍色 (B) 通道的強度
 b. 降低紅色 (R) 通道的強度

```python
b_up = b + 30 # Increase the intensity of the blue (B) channel
r_down = r // 2 # Decrease the intensity of the red (R) channel
```

5. 合併顏色通道。

`cv2.merge()` 函數將 3 個輸入顏色通道合併為單一彩色影像。

建立 2 個具有冷色調效果的影像：增加藍色強度的「img_b_up」和減少紅色強度的「img_r_down」。

```python
img_b_up = cv2.merge([r, g, b_up])
img_r_down = cv2.merge([r_down, g, b])
```

6. 顯示並比較處理後的結果影像。

`plt.subplots_adjust()` 函數可讓您調整圖中子圖（圖像）之間的間距。參數「wspace」專門控制子圖之間的水平空間。

在顯示多個圖像之前，您需要一個畫布或框架來放置它們。 `plt.subplots()` 函數為您建立這個畫布，也稱為圖形。

`figsize` 參數決定了該圖形的大小。它採用兩個值的元組，表示圖形的寬度和高度（以英吋為單位）。

`set_title()` 用於為每個子圖新增標題。

```python
plt.subplots_adjust(wspace = 0.6)
plt.subplots(figsize = (10, 5))
plt.axis('off')  # Remove the axes

ax1 = plt.subplot(1, 3, 1)
ax1.set_title("Original Image") # Display the title
ax1.axis('off')
plt.imshow(img, vmin = 0, vmax = 255)

ax2 = plt.subplot(1, 3, 2)
ax2.set_title("Increased Blue Intensity") # Display the title
ax2.axis('off')
plt.imshow(img_b_up, vmin = 0, vmax = 255)

ax3 = plt.subplot(1, 3, 3)
ax3.set_title("Decreased Red Intensity") # Display the title
ax3.axis('off')
plt.imshow(img_r_down, vmin = 0, vmax = 255)

plt.show()
```

### **[教師示範] 影像的直方圖均衡化**

1. 讀取並顯示彩色影像`yh_worse.jpg`。

影像的對比度很差，這意味著整個影像的光強度沒有太大變化。影像均衡可以幫助解決這個問題。

```python
img_w = cv2.imread("/content/Computer_Vision/First_Lesson/images/yh_worse.jpg")[:, :, ::-1]

plt.imshow(img_w, vmin = 0, vmax = 255)
plt.show()
```

2. 將影像分割為 3 個色彩通道。

```python
r, g, b = cv2.split(img_w)
```

3. 對每個顏色通道執行直方圖均衡。

```python
rH = cv2.equalizeHist(r)
gH = cv2.equalizeHist(g)
bH = cv2.equalizeHist(b)
```

4. 合併顏色通道。

```python
img_h = cv2.merge([rH, gH, bH])
```

5. 顯示並比較處理後的結果影像。

```python
ax1 = plt.subplot(1, 2, 1)
ax1.set_title("Original Image")
ax1.axis('off')
plt.imshow(img_w, vmin = 0, vmax = 255)

ax2 = plt.subplot(1, 2, 2)
ax2.set_title("Processed Image")
ax2.axis('off')
plt.imshow(img_h, vmin = 0, vmax = 255)
```
